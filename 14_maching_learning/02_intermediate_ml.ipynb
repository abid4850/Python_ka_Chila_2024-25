```python
# write the whole script in one go
# import libraries
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LinearRegression
# load the data
df = sns.load_dataset('tips')
# select the feature column from df
X = df[['total_bill']]
# select the target column from df
y = df['tip']
# load/call the linear regression model to
model = LinearRegression()
# fit the model means train the model
model.fit(X,y)
# lets predict the tip for an unknown total_bill
model.predict([[40]])
```

```python
# Create a regression line plot
sns.lmplot(x='total_bill', y='tip', data=df)
```

```python
# write the whole script in one go
# import libraries
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LinearRegression
# load the data
df = sns.load_dataset('tips')
# select the feature column from df
X = df[['total_bill', 'size']]
# select the target column from df
y = df['tip']
# load/call the linear regression model to
model = LinearRegression()
# fit the model means train the model
model.fit(X,y)
# lets predict the tip for an unknown total_bill
model.predict([[35.574,  2]])
```

```python
# write the whole script in one go
# import libraries
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LinearRegression
# load the data
df = sns.load_dataset('tips')
# encode smoker columns 
df['smoker_num']= df['smoker'].map({'Yes':1, 'No':0})
df.head()
# select the feature column from df
X = df[['total_bill', 'size', 'smoker_num']]
# select the target column from df
y = df['tip']
# load/call the linear regression model to
model = LinearRegression()
# fit the model means train the model
model.fit(X,y)
# lets predict the tip for an unknown total_bill
print(model.predict([[35.574, 3, 1]]))
print(model.predict([[35.574, 3, 0]]))
```

# Classification
### Logistic Regression

```python
# write the whole script in one go
# import libraries
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LinearRegression, LogisticRegression
# load the data
df = sns.load_dataset('tips')
# encode smoker columns 
df['smoker_num']= df['smoker'].map({'Yes':1, 'No':0})
df.head()
# select the feature column from df
X = df[['total_bill', 'size', 'tip']]
# select the target column from df
y = df['smoker']
# load/call the linear regression model to
model = LogisticRegression()
# fit the model means train the model
model.fit(X,y)
# lets predict 
model.predict([[50, 3, 5.67]])
```
